```python
import matplotlib.pyplot as plt

nomes = ['rafael', 'Flavio', 'Maria']
idades = [10,20,15]
```

```python
plt.plot(nomes,idades)
```

```python
plt.figure(figsize=(5,2)) # Tamanho largura e altura do grafico
plt.bar(nomes,idades) # Grafico em barra

plt.xlabel('nome') # Legenda nos graficos
plt.ylabel('idade') 
```
